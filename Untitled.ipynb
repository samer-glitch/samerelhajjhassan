```python
import matplotlib.lines as mlines
import matplotlib.pyplot as plt
from matplotlib.ticker import (
    MultipleLocator, 
    StrMethodFormatter
)
import numpy as np

class dist:
    scales = np.array([2, 4])
    factors = np.array([49/198, 25/198])
    @classmethod
    def pdf(cls, x):
        return sum(
            np.exp(x * (-1) / scale / 60) * factor / 60 for scale, factor in zip(
                cls.scales, 
                cls.factors
            )
        )
    @classmethod
    def ppf(cls, q):
        return 60 * 4 * np.log( -49 / ( 25 - np.sqrt( 5476 - 4851 * q ) ) )
    @classmethod
    def mean(cls):
        return sum( cls.factors * cls.scales ** 2 ) * 60
    @classmethod
    def var(cls):
        return  2 * sum( cls.factors * ( cls.scales ** 3 ) ) * 60 ** 2 - cls.mean() ** 2 
    @classmethod
    def rvs(cls, size):
        return cls.ppf(np.random.uniform(size=size))

mean = dist.mean()
var = dist.var()
std = np.sqrt(var)
quartiles = np.array(list(dist.ppf(.25 * i) for i in range(1, 4)))
x = np.arange(
    0,
    max(int(dist.ppf(.99)), int(var)) + 60, 
    10
)
y = dist.pdf(x)

fig, ax = plt.subplots(2, figsize=(10, 8), gridspec_kw={'height_ratios': [3, 2]})
ax[0].plot(x, y, 'bo-', ms=4, lw=5, alpha=0.5, label='pdf')
ax[0].vlines(quartiles, 0, dist.pdf(quartiles), colors='m',  ls='--', lw=4, alpha=0.5, label='_quartiles')
ax[0].plot(quartiles, dist.pdf(quartiles), 'mo', ms=10, alpha=0.5, label='_quartiles')
ax[0].vlines(mean, 0, dist.pdf(mean), colors='r',  ls=':', lw=4, alpha=0.5, label='_mean')
ax[0].plot(mean, dist.pdf(mean), 'ro', ms=14, alpha=0.5, label='_mean')
ax[0].vlines(std, 0, dist.pdf(std), colors='g',  ls=':', lw=4, alpha=0.5, label='_std')
ax[0].plot(std, dist.pdf(std), 'go', ms=14, alpha=0.5, label='_std')
x_sub = np.arange(120, 240 + 1, 30)
ax[0].set_title('Probability density function', fontsize=16)
ax[0].grid(True, which='both')
handles, labels = ax[0].get_legend_handles_labels()
handles.extend(
    [
        mlines.Line2D([], [], color='m', marker='o', ls=':', ms=10, lw=4, alpha=.5),
        mlines.Line2D([], [], color='r', marker='o', ls=':', ms=14, lw=4, alpha=.5),
        mlines.Line2D([], [], color='g', marker='o', ls=':', ms=14, lw=4, alpha=.5)
    ]
)
labels.extend(['quartiles','mean','standard deviation'])
ax[0].legend(handles=handles, labels=labels)
ax[0].xaxis.set_major_locator(MultipleLocator(60))
ax[0].xaxis.set_minor_locator(MultipleLocator(10))
ax[0].tick_params(axis='x', labelrotation=45)
ax[0].yaxis.set_major_formatter(StrMethodFormatter('{x:.2g}'))
ax[0].set(xlim=(0, max([std, mean, *quartiles]) * 2), ylim=(0, y.max() * 1.05), autoscale_on=False)
for point in (mean, std, *quartiles):
    ax[0].annotate(
        f"{point:.2f}", 
        xy=(point, dist.pdf(point)),  
        xycoords='data',
        xytext=(point, dist.pdf(point) * 1.1), 
        textcoords='data'
    )
sample_size = 10_000
random_sample = dist.rvs(sample_size)
nbins = int(np.ceil(np.log2(sample_size))) + 1
equiareal = np.interp(np.linspace(0, sample_size, nbins + 1), np.arange(sample_size), np.sort(random_sample))
n, bins, patches = ax[1].hist(random_sample, bins=equiareal, density=True, edgecolor='black', alpha=.3)
mid = .5*(bins[1:] + bins[:-1])
ax[1].errorbar(mid, n, yerr=n ** 1.2, fmt='none', ecolor='black')
ax[1].set(xlim=(0, max([std, mean, *quartiles]) * 2), ylim=(0,None), autoscale_on=True)
ax[1].xaxis.set_major_locator(MultipleLocator(60))
ax[1].xaxis.set_minor_locator(MultipleLocator(10))
ax[1].set_title(f'Histogram of random sample (n={sample_size:,})', fontsize=16)
ax[1].yaxis.set_major_formatter(StrMethodFormatter('{x:.2g}'))
fig.suptitle('Time to hear an owl', fontsize=18)
fig.supxlabel('Time [min]', fontsize=14)
fig.supylabel('Probability density', fontsize=14)
fig.tight_layout()
fig.savefig("1_3.svg", bbox_inches="tight")
plt.show()
```
