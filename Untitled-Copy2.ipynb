```python
import matplotlib.lines as mlines
import matplotlib.pyplot as plt
from matplotlib.ticker import (
    MultipleLocator, 
    StrMethodFormatter
)
import numpy as np

class dist:
    scales = np.array([2, 4])
    factors = np.array([49/198, 25/198])
    @classmethod
    def pdf(cls, x):
        return sum(
            np.exp(x * (-1) / scale / 60) * factor / 60 for scale, factor in zip(
                cls.scales, 
                cls.factors
            )
        )
    @classmethod
    def ppf(cls, q):
        return 60 * 4 * np.log( -49 / ( 25 - np.sqrt( 5476 - 4851 * q ) ) )
    @classmethod
    def mean(cls):
        return sum( cls.factors * cls.scales ** 2 ) * 60
    @classmethod
    def var(cls):
        return  2 * sum( cls.factors * ( cls.scales ** 3 ) ) * 60 ** 2 - cls.mean() ** 2 
    @classmethod
    def rvs(cls, size):
        return cls.ppf(np.random.uniform(size=size))

mean = dist.mean()
var = dist.var()
std = np.sqrt(var)
quartiles = np.array(list(dist.ppf(.25 * i) for i in range(1, 4)))
x = np.arange(
    0,
    max(int(dist.ppf(.99)), int(var)) + 60, 
    10
)
y = dist.pdf(x)

fig, ax = plt.subplots(2, figsize=(10, 8), gridspec_kw={'height_ratios': [3, 2]})
ax[0].plot(x, y, 'bo-', ms=4, lw=5, alpha=0.5, label='pdf')
ax[0].vlines(quartiles, 0, dist.pdf(quartiles), colors='m',  ls='--', lw=4, alpha=0.5, label='_quartiles')
ax[0].plot(quartiles, dist.pdf(quartiles), 'mo', ms=10, alpha=0.5, label='_quartiles')
ax[0].vlines(mean, 0, dist.pdf(mean), colors='r',  ls=':', lw=4, alpha=0.5, label='_mean')
ax[0].plot(mean, dist.pdf(mean), 'ro', ms=14, alpha=0.5, label='_mean')
ax[0].vlines(std, 0, dist.pdf(std), colors='g',  ls=':', lw=4, alpha=0.5, label='_std')
ax[0].plot(std, dist.pdf(std), 'go', ms=14, alpha=0.5, label='_std')
x_sub = np.arange(120, 240 + 1, 30)
ax[0].set_title('Probability density function', fontsize=16)
ax[0].grid(True, which='both')
handles, labels = ax[0].get_legend_handles_labels()
handles.extend(
    [
        mlines.Line2D([], [], color='m', marker='o', ls=':', ms=10, lw=4, alpha=.5),
        mlines.Line2D([], [], color='r', marker='o', ls=':', ms=14, lw=4, alpha=.5),
        mlines.Line2D([], [], color='g', marker='o', ls=':', ms=14, lw=4, alpha=.5)
    ]
)
labels.extend(['quartiles','mean','standard deviation'])
ax[0].legend(handles=handles, labels=labels)
ax[0].xaxis.set_major_locator(MultipleLocator(60))
ax[0].xaxis.set_minor_locator(MultipleLocator(10))
ax[0].tick_params(axis='x', labelrotation=45)
ax[0].yaxis.set_major_formatter(StrMethodFormatter('{x:.2g}'))
ax[0].set(xlim=(0, max([std, mean, *quartiles]) * 2), ylim=(0, y.max() * 1.05), autoscale_on=False)
for point in (mean, std, *quartiles):
    ax[0].annotate(
        f"{point:.2f}", 
        xy=(point, dist.pdf(point)),  
        xycoords='data',
        xytext=(point, dist.pdf(point) * 1.1), 
        textcoords='data'
    )
sample_size = 10_000
random_sample = dist.rvs(sample_size)
nbins = int(np.ceil(np.log2(sample_size))) + 1
equiareal = np.interp(np.linspace(0, sample_size, nbins + 1), np.arange(sample_size), np.sort(random_sample))
n, bins, patches = ax[1].hist(random_sample, bins=equiareal, density=True, edgecolor='black', alpha=.3)
mid = .5*(bins[1:] + bins[:-1])
ax[1].errorbar(mid, n, yerr=n ** 1.2, fmt='none', ecolor='black')
ax[1].set(xlim=(0, max([std, mean, *quartiles]) * 2), ylim=(0,None), autoscale_on=True)
ax[1].xaxis.set_major_locator(MultipleLocator(60))
ax[1].xaxis.set_minor_locator(MultipleLocator(10))
ax[1].set_title(f'Histogram of random sample (n={sample_size:,})', fontsize=16)
ax[1].yaxis.set_major_formatter(StrMethodFormatter('{x:.2g}'))
fig.suptitle('Time to hear an owl', fontsize=18)
fig.supxlabel('Time [min]', fontsize=14)
fig.supylabel('Probability density', fontsize=14)
fig.tight_layout()
fig.savefig("1_3.svg", bbox_inches="tight")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from scipy import optimize

def f(y):
    return (28/99) * np.exp(-8*y**2) + (71/99)* np.exp(-5*y**2)

mean, _ = quad(lambda y: y*f(y), 0, np.inf)
variance, _ = quad(lambda y: (y-mean)**2*f(y), 0, np.inf)

print(f"Mean:, {mean:.3f}")
print(f"Variance:, {variance:.3f}")

# Define a function to calculate the integral of f(y) from 0 to y
def F(y):
    return quad(f, 0, y)[0]

# Find the quartiles by solving for the values of y where F(y)/F(inf) is 0.25, 0.5, and 0.75
# Using the quad function from the scipy.integrate module to numerically integrate the function f(y) from 0 to infinity. The result of this integration is assigned to the variable F_inf.

F_inf = quad(f, 0, np.inf)[0]

# The line q1 = optimize.brentq(lambda y: F(y)/F_inf - 0.25, 0, 10) is using the brentq function from the scipy.optimize module to find the root of the lambda function lambda y: F(y)/F_inf - 0.25 in the interval from 0 to 10. The root of this function is the value of y that satisfies the equation ∫[0, y] f(y)dy = 0.25. The value of this root is assigned to the variable q1.
q1 = optimize.brentq(lambda y: F(y)/F_inf - 0.25, 0, 10)
q2 = optimize.brentq(lambda y: F(y)/F_inf - 0.5, 0, 10)
q3 = optimize.brentq(lambda y: F(y)/F_inf - 0.75, 0, 10)


y = np.linspace(0, 5, 1000)  # in hours
pdf = f(y)  # probability density function
bins = np.arange(0, 5*60+1, 1)  # in minutes
hist, _ = np.histogram(np.random.choice(y, size=int(1e6), p=pdf/np.sum(pdf))*60, bins=bins, density=True)

print(f"Quartile 1 or 25th percentile:, {q1:.3f}")
print(f"Quartile 2 or 50th percentile:, {q2:.3f}")
print(f"Quartile 3 or 75th percentile:, {q3:.3f}")
plt.axvline(mean, color="r", linestyle="-", label="Mean")
plt.axvline(q1, color="g", linestyle="-.", label="Q1")
plt.axvline(q2, color="y", linestyle="-.", label="Median")
plt.axvline(q3, color="r", linestyle="-.", label="Q3")
#plt.plot(y, pdf, label="")
plt.plot(y, pdf, label="Probability Density Function")
plt.ylabel("Probability")
plt.title("Waiting Time to Hear an Owl")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the probability density function
def pdf(y):
    return (28/99)*np.exp(-8*y**2) + (71/99)*np.exp(-5*y**2)

# Set the random seed for reproducibility
np.random.seed(42)

# Generate a sample of 10000 values from the PDF over a range of 3 hours
y_values = np.linspace(0, 3, 1000)
probs = pdf(y_values) / np.sum(pdf(y_values))
y_sample = np.random.choice(y_values, size=100000, p=probs)

# Convert the sample from hours to minutes
y_sample = y_sample * 60

# Compute the frequency of events per minute
counts, bins = np.histogram(y_sample, bins=180, range=(0, 180))
freq = counts / 180

# Plot the histogram on a graph
fig, ax = plt.subplots()
ax.bar(bins[:-1], freq, width=1)
ax.set_xlabel('Time (minutes)')
ax.set_ylabel('Frequency')
ax.set_title('Frequency of Events per Minute')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the probability density function
def pdf(y):
    return (28/99)*np.exp(-8*y**2) + (71/99)*np.exp(-5*y**2)

# Set the random seed for reproducibility
np.random.seed(42)

# Generate a sample of 10000 values from the PDF over a range of 180 minutes
y_values = np.linspace(0, 3, 180)
probs = pdf(y_values) / np.sum(pdf(y_values))
y_sample = np.random.choice(y_values, size=10000, p=probs)

# Convert the sample from hours to minutes
y_sample = y_sample * 60

# Compute the histogram using probabilities
counts, bins = np.histogram(y_sample, bins=180, range=(0, 180), density=True)

# Compute the midpoints of the bins
bin_midpoints = (bins[1:] + bins[:-1]) / 2

# Plot the histogram with bars and lines connecting the midpoints
fig, ax = plt.subplots()
ax.bar(bins[:-1], counts, width=1, alpha=0.5)
ax.plot(bin_midpoints, counts, 'r.-', linewidth=2)
ax.set_xlabel('Time (minutes)')
ax.set_ylabel('Probability density')
ax.set_title('Histogram with Bars and Lines')
plt.show()
```
