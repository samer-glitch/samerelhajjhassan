```python
import numpy as np
import matplotlib.pyplot as plt

observations = [303, 125, 31, 31, 140]

def pdf(t, theta):
    if t < 0:
        return 0
    elif t <= theta:
        return t / (theta ** 2)
    else:
        return 0

t_values = np.linspace(0, max(observations), 100)
theta_values = np.linspace(0, max(observations), 100)

likelihood_values = []
for theta in theta_values:
    likelihood = np.prod([pdf(t, theta) for t in observations])
    likelihood_values.append(likelihood)

mle_index = np.argmax(likelihood_values)
mle_theta = theta_values[mle_index]

plt.figure(figsize=(10, 5))

# Plot the PDF
plt.subplot(1, 2, 1)
plt.plot(t_values, [pdf(t, mle_theta) for t in t_values], label=f't/θ^2 (MLE: θ={mle_theta:.2f})')
plt.scatter(observations, [0] * len(observations), color='red', label='Observations')
plt.xlabel('t')
plt.ylabel('PDF')
plt.legend()

# Plot the likelihood function
plt.subplot(1, 2, 2)
plt.plot(theta_values, likelihood_values)
plt.scatter(mle_theta, max(likelihood_values), color='red', label='MLE')
plt.xlabel('θ')
plt.ylabel('Likelihood')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import sympy as sp

observations = [303, 125, 31, 31, 140]
theta = sp.symbols('theta')

# Define the joint function
L = sp.prod((-t + 2*theta) / theta**2 for t in observations)

# Differentiate the joint function with respect to theta
dL = sp.diff(L, theta)

# Solve for the MLE value(s) of theta
mle_theta = sp.solve(dL, theta)

# Print the MLE value(s)
print("MLE for theta:")
for theta_value in mle_theta:
    print(theta_value.evalf())
```

```python
import numpy as np
import matplotlib.pyplot as plt

observations = [303, 125, 31, 31, 140]

def pdf(t, theta):
    if t < theta or t > 2*theta:
        return 0
    else:
        return (-t + 2*theta) / (theta**2)

theta_values = np.linspace(0, max(observations), 100)

likelihood_values = []
for theta in theta_values:
    likelihood = np.prod([pdf(t, theta) for t in observations])
    likelihood_values.append(likelihood)

mle_index = np.argmax(likelihood_values)
mle_theta = theta_values[mle_index]

plt.plot(theta_values, likelihood_values)
plt.scatter(mle_theta, max(likelihood_values), color='red', label='MLE')
plt.xlabel('θ')
plt.ylabel('Likelihood')
plt.title('Likelihood Function')
plt.legend()
plt.show()

print(f"Maximum Likelihood Estimator (MLE): θ = {mle_theta:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.ticker import MultipleLocator

t = np.array([303, 125, 31, 31, 140])
def pdf(t, theta):
    return (-t) / (theta**2)
def likelihood(t, theta):
    try: iter(theta)
    except: theta = [theta]
    return np.array([np.prod(pdf(t, val)) for val in theta])
theta = np.arange(.1, 400.1, .1)
y = likelihood(t, theta)
highest_likelihood = y.max()
optimal_theta = theta[np.where(y == highest_likelihood)[0][0]]

fig, ax = plt.subplots(2, figsize=(10, 8), gridspec_kw={'height_ratios': [2, 2]})
ax[0].plot(theta, y, 'b-', lw=2, alpha=0.5, label='likelihood')
ax[0].plot(20, likelihood(t, 20), 'bo', ms=10, alpha=0.5, label='$\\theta = 20$')
ax[0].plot(optimal_theta, highest_likelihood, 'mo', ms=10, alpha=0.5, label=f'$\\theta = {optimal_theta:.1f}$')
ax[0].plot(40, likelihood(t, 40), 'ko', ms=10, alpha=0.5, label='$\\theta = 40$')
ax[0].set_title(f'Likelihood function for the sample t=[{",".join(map(str, t))}]', fontsize=16)
ax[0].set_xlabel(f'Scale parameter $\\theta$', fontsize=14)
ax[0].set_ylabel('$\\mathcal{L}(\\theta)$', fontsize=14)
ax[0].grid(True, which='both')
ax[0].legend()
ax[0].xaxis.set_major_locator(MultipleLocator(10))
ax[0].xaxis.set_minor_locator(MultipleLocator(1))
ax[0].set(xlim=(0, theta[-1]), ylim=(0, highest_likelihood * 1.2), autoscale_on=False)
for theta, color in zip((20, optimal_theta, 40), ('b', 'm', 'k')):
    res = likelihood(t, theta)[0]
    ax[0].annotate(
        f"{res:.5g}", 
        xy=(theta, res),  
        xycoords='data',
        xytext=(theta - .4*len(f"{res:.5g}"), res * 1.07), 
        textcoords='data',
        color=color
    )
sample = np.linspace(0, 300, 100)
for theta, c, ls, lw in zip((20, optimal_theta, 40), ('k', 'b', 'k'), ('--', '-', ':'), (2, 2, 3)):
    ax[1].plot(sample, pdf(sample, theta), f'{c}{ls}', lw=lw, alpha=0.5, label='_sample1')
    ax[1].plot(t, pdf(t, theta), f'{c}o', ms=7, alpha=0.5, label='_sample2')
    ax[1].vlines(2 * theta, 0, pdf(2 * theta, theta), colors=c,  ls=ls, lw=lw*2, alpha=0.5, label='_mean')
handles, labels = [
    mlines.Line2D([], [], color='k', marker='o', ls='--', ms=7, lw=2, alpha=.5),
    mlines.Line2D([], [], color='b', marker='o', ls='-', ms=7, lw=2, alpha=.5),
    mlines.Line2D([], [], color='k', marker='o', ls=':', ms=7, lw=2, alpha=.5)
], ['$\\theta = 20$',f'$\\theta = {optimal_theta:.1f}$','$\\theta = 40$']
ax[1].legend(handles=handles, labels=labels)
ax[1].set_title(f'Probability density functions for $\\theta = [20,{optimal_theta:.1f},40]$', fontsize=16)
ax[1].xaxis.set_major_locator(MultipleLocator(10))
ax[1].xaxis.set_minor_locator(MultipleLocator(5))
ax[1].set(xlim=(0, sample[-1]), ylim=(0, 0.02), autoscale_on=False)
ax[1].set_xlabel('Sample $t$', fontsize=14)
ax[1].set_ylabel('Probability density $s_T(t|\\theta)$', fontsize=14)
ax[1].grid(True, which='major')
fig.suptitle('Failure of dual routers', fontsize=18)
fig.tight_layout()
fig.savefig("3.svg", bbox_inches="tight")
plt.show()
```
