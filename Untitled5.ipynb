```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter θ
theta = 100

# Range of values for T
t_values = np.linspace(0, theta, 1000)

# Calculate the PDF of T
pdf_t = t_values / (theta**2)

# Plot the PDF of T
plt.plot(t_values, pdf_t, label='PDF of T')
plt.xlabel('T')
plt.ylabel('Density')
plt.title('Probability Density Function of T')
plt.legend()
plt.show()

# Given sample
sample = [303, 125, 31, 31, 140]

# Calculate the likelihood function for θ
likelihood = np.prod(sample) / (theta**10)

# Plot the likelihood function
plt.plot(theta, likelihood, 'ro', label='Likelihood')
plt.xlabel('θ')
plt.ylabel('Likelihood')
plt.title('Likelihood Function')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.ticker import MultipleLocator

t = np.array([303, 125, 31, 31, 140])
def pdf(t, theta):
    return t / (theta ** 2) 
def likelihood(t, theta):
    try: iter(theta)
    except: theta = [theta]
    return np.array([np.prod(pdf(t, val)) for val in theta])
theta = np.arange(1, 20.1, 1)
y = likelihood(t, theta)
highest_likelihood = y.max()
optimal_theta = theta[np.where(y == highest_likelihood)[0][0]]

fig, ax = plt.subplots(2, figsize=(10, 8), gridspec_kw={'height_ratios': [2, 2]})
ax[0].plot(theta, y, 'b-', lw=2, alpha=0.5, label='likelihood')
ax[0].plot(5, likelihood(t, 5), 'bo', ms=10, alpha=0.5, label='$\\theta = 20$')
ax[0].plot(optimal_theta, highest_likelihood, 'mo', ms=10, alpha=0.5, label=f'$\\theta = {optimal_theta:.1f}$')
ax[0].plot(3, likelihood(t, 3), 'ko', ms=10, alpha=0.5, label='$\\theta = 20$')
ax[0].set_title(f'Likelihood function for the sample t=[{",".join(map(str, t))}]', fontsize=16)
ax[0].set_xlabel(f'Scale parameter $\\theta$', fontsize=14)
ax[0].set_ylabel('$\\mathcal{L}(\\theta)$', fontsize=14)
ax[0].grid(True, which='both')
ax[0].legend()
ax[0].xaxis.set_major_locator(MultipleLocator(10))
ax[0].xaxis.set_minor_locator(MultipleLocator(1))
ax[0].set(xlim=(0, theta[-1]), ylim=(0, highest_likelihood * 1.2), autoscale_on=False)
for theta, color in zip((20, optimal_theta, 40), ('b', 'm', 'k')):
    res = likelihood(t, theta)[0]
    ax[0].annotate(
        f"{res:.5g}", 
        xy=(theta, res),  
        xycoords='data',
        xytext=(theta - .4*len(f"{res:.5g}"), res * 1.07), 
        textcoords='data',
        color=color
    )
sample = np.linspace(0, 20, 100)
for theta, c, ls, lw in zip((20, optimal_theta, 40), ('k', 'b', 'k'), ('--', '-', ':'), (2, 2, 3)):
    ax[1].plot(sample, pdf(sample, theta), f'{c}{ls}', lw=lw, alpha=0.5, label='_sample1')
    ax[1].plot(t, pdf(t, theta), f'{c}o', ms=7, alpha=0.5, label='_sample2')
    ax[1].vlines(2 * theta, 0, pdf(2 * theta, theta), colors=c,  ls=ls, lw=lw*2, alpha=0.5, label='_mean')
handles, labels = [
    mlines.Line2D([], [], color='k', marker='o', ls='--', ms=7, lw=2, alpha=.5),
    mlines.Line2D([], [], color='b', marker='o', ls='-', ms=7, lw=2, alpha=.5),
    mlines.Line2D([], [], color='k', marker='o', ls=':', ms=7, lw=2, alpha=.5)
], ['$\\theta = 20$',f'$\\theta = {optimal_theta:.1f}$','$\\theta = 40$']
ax[1].legend(handles=handles, labels=labels)
ax[1].set_title(f'Probability density functions for $\\theta = [20,{optimal_theta:.1f},40]$', fontsize=16)
ax[1].xaxis.set_major_locator(MultipleLocator(10))
ax[1].xaxis.set_minor_locator(MultipleLocator(5))
ax[1].set(xlim=(0, sample[-1]), ylim=(0, 0.02), autoscale_on=False)
ax[1].set_xlabel('Sample $t$', fontsize=14)
ax[1].set_ylabel('Probability density $s_T(t|\\theta)$', fontsize=14)
ax[1].grid(True, which='major')
fig.suptitle('Failure of dual routers', fontsize=18)
fig.tight_layout()
fig.savefig("3.svg", bbox_inches="tight")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

# Define the probability density function
def pdf(y):return (448/99)*y*np.exp(-8*y**2) + (710/99)*y*np.exp(-5*y**2)

# Define the cumulative distribution function
def cdf(y):return 1 - (28/99)*np.exp(-8*y**2) - (71/99)*np.exp(-5*y**2)

# Define the range of y values
y_values = np.linspace(0, 5, 1000)

# Compute the pdf and cdf values for the range of y values
pdf_values = pdf(y_values)
cdf_values = cdf(y_values)

# Calculate the probability between 2 and 4 hours using integration
probability_2_to_4 = cdf(4) - cdf(2)
print("The probability that you need to wait between 2 and 4 hours to hear the owl = ", probability_2_to_4)

# Compute the mean, variance, and quartiles using numerical integration
mean, _ = integrate.quad(lambda y: y*pdf(y), 0, np.inf)
variance, _ = integrate.quad(lambda y: (y-mean)**2*pdf(y), 0, np.inf)

# Print the mean and variance
print(f"Mean: {mean:.2f} hours = {mean*60:.2f} minutes")
print(f"Variance: {variance:.2f} hours\u00B2 = {variance*60:.2f} minutes\u00B2")

# Find the quartiles of the CDF
F_inf = integrate.quad(pdf, 0, np.inf)[0]
#print("F_inf", F_inf)
q1 = 0
q2 = 0
q3 = 0
for y in np.linspace(0, 3, 10000): # the iterations takes place from 0 to 3 hours
    F_y = cdf(y) / F_inf
    if F_y >= 0.25 and q1 == 0:
        q1 = y
    if F_y >= 0.5 and q2 == 0:
        q2 = y
    if F_y >= 0.75 and q3 == 0:
        q3 = y
    if q1 > 0 and q2 > 0 and q3 > 0:
        break

# Format the output to make Q1, Q2, and Q3 bold
print(f"\nFirst quartile           Q1: {q1:.4f} hours = {q1*60:.4f} minutes")
print(f"Second quartile (median) Q2: {q2:.4f} hours = {q2*60:.4f} minutes")
print(f"Third quartile           Q3: {q3:.4f} hours = {q3*60:.4f} minutes\n")

# Plot the pdf and cdf
fig, ax = plt.subplots(figsize=(10, 6))  
ax.plot(y_values, pdf_values, label='pdf')
ax.plot(y_values, cdf_values, label='cdf')
ax.grid(True)
ax.set_xlabel('Time (hours)')
ax.set_ylabel('Probability density')
ax.set_title('PDF and CDF of Waiting Times to Hear an Owl')
ax.legend()

# Add vertical dashed lines for the quartiles
ax.axvline(x=q1, linestyle='--', color='r', label='Q1')
ax.axvline(x=q2, linestyle='--', color='g', label='Q2 (Median)')
ax.axvline(x=q3, linestyle='--', color='b', label='Q3')

# Add arrows pointing to the quartiles
arrow_props = dict(facecolor='black', arrowstyle='->')
ax.annotate(r'$\bf{Q1}$', xy=(q1, 0), xytext=(q1, 0.1),
            arrowprops=arrow_props, ha='center')
ax.annotate(r'$\bf{Q2}$', xy=(q2, 0), xytext=(q2, 0.2),
            arrowprops=arrow_props, ha='center')
ax.annotate(r'$\bf{Q3}$', xy=(q3, 0), xytext=(q3, 0.06),
            arrowprops=arrow_props, ha='center')

ax.legend()
plt.show()

# Another Method using Trapz function
# Compute and display mean, variance, and quartiles
#mean = np.trapz(y_values * pdf_values, y_values)
#variance = np.trapz((y_values - mean)**2 * pdf_values, y_values)
#q11 = np.interp(0.25, cdf_values, y_values)
#q22 = np.interp(0.5, cdf_values, y_values)
#q33 = np.interp(0.75, cdf_values, y_values)

# Format the output to make Q1, Q2, and Q3 bold
#print(f"Another Approach to get the Quartiles")
#print(f"First quartile           Q1: {q11:.4f} hours = {q1*60:.4f} minutes")
#print(f"Second quartile (median) Q2: {q22:.4f} hours = {q2*60:.4f} minutes")
#print(f"Third quartile           Q3: {q33:.4f} hours = {q3*60:.4f} minutes")
```

```python
import numpy as np
import matplotlib.pyplot as plt
def s(t, θ):
    if t < 0:
        return 0
    elif 0 <= t <= θ:
        return t / θ**2
    elif θ <= t <= 2*θ:
        return (-t) / θ**2 + 2/θ
    else:
        return 0
# Values of θ
θ_values = [0.5, 1, 1.5]
# Time values
t = np.linspace(0, 2*max(θ_values), 500)
# Plotting
plt.figure(figsize=(8, 6))
for θ in θ_values:
    s_values = [s(time, θ) for time in t]
    plt.plot(t, s_values, label=f'θ = {θ}')
# Add 1/θ on y-axis
for θ in θ_values:
    plt.axhline(y=1/θ, linestyle='--',
color='gray', alpha=0.5)
plt.text(t[-1], 1/θ_values[0], '1/θ', ha='right',
va='center', color='gray')
plt.xlabel('Time (t)')
plt.ylabel('s(t)')
plt.title('Plot of s(t) with respect to θ')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math
# Define the sample data
T = [303, 125, 31, 31, 140]
# Define the log-likelihood function
def log_likelihood(theta):
    if all(0 < t < theta for t in T): # Check if the data is in the first interval
        return sum(math.log(t / theta**2) for t in T) # Return the log-likelihood value
    elif all(theta <= t <= 2*theta for t in T): # Check if the data is in the second interval
        return sum((-t) / theta**2 + theta/2 for t in T) # Return the log-likelihood value
    else:
        return -math.inf # Return a very small value otherwise
# Find the MLE of theta by maximizing the log-likelihood function
MLE = max(T) # The MLE is equal to the maximum value in the data
print("The MLE of theta is", MLE)
```
